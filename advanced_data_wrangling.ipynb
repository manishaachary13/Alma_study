```python
# import all necessary modules here
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
path='/content/drive/MyDrive/Colab Notebooks/imdb_data.csv'
# read csv file and store in a dataframe
imdb_df=pd.read_csv(path)
```

```python
# convert title into uppercase
def to_uppercase(column):
    return column.upper()
# apply metyhod call a function which does the actual operation.
imdb_df['capitalized_title']=imdb_df['title'].apply(to_uppercase)
# print the column
imdb_df['capitalized_title']
```

```python
# use lambda function to calculate profit out of revenue and budget
# this function will return profit for each row data
def revised_profit(revenue,budget):
    if revenue > 0:
        return revenue - budget
    else:
        return np.nan
#here apply function call lambda function which call revised_profit function to return profit for each data
imdb_df['new_profit']=imdb_df.apply(lambda x: revised_profit(x['revenue'],x['budget']),axis=1)
# printing new profit
imdb_df['new_profit']
```

We can merge dataframes using the following methods

* <code> pd.concat( [df] , axis) </code>

* <code> pd.merge( df1, df2 ,on,how,suffixes) </code>

```python
# let's create 3 datasets to implement these methods
# divide runtime into 3 categories according to some range condition
imdb1=imdb_df[imdb_df['runtime']>150]
imdb2=imdb_df[(imdb_df['runtime'] < 150) & (imdb_df['runtime']>100)]
imdb3=imdb_df[imdb_df['runtime']<100]
```

```python
# concatenate these 3 datasets
concate_data=pd.concat([imdb1,imdb2,imdb3])
# .shape will print how many rows and columns the dataframe have
concate_data.shape
```

```python
# create a dataframe which consist of only english language
english=imdb_df[imdb_df['original_language'] == 'en']
# create another language whose runtime is more than 100
long_movies=imdb_df.loc[imdb_df['runtime']>100, ['imdb_id','original_language']]
# compare shape of both dataframe
english.shape, long_movies.shape
```

```python
# merging two dataframes using inner join
inner_merge=pd.merge(english, long_movies, how='inner',left_on='imdb_id',right_on='imdb_id')
inner_merge
```

```python
# left join two dataframe
left_merge=pd.merge(english, long_movies, how="left",left_on='imdb_id',right_on='imdb_id')
left_merge
```

```python
right_merge=pd.merge(english, long_movies, how="right",left_on='imdb_id',right_on='imdb_id')
right_merge
```

```python
# Outer Join
merged_OJ_df = pd.merge(english[['imdb_id','runtime']],long_movies, on='imdb_id', how='outer')
```

```python
# Convert the "Release Date" column to datetime type if it's not already in datetime format
imdb_df['release_date'] = pd.to_datetime(imdb_df['release_date'], errors='coerce')

# Extract the release year and add it as a new column
imdb_df['release_year'] = imdb_df['release_date'].dt.year
# We can find the maximum movie runtime across different years
max_runtime_movies_by_year = imdb_df.groupby('release_year')['runtime'].max().reset_index()
max_runtime_movies_by_year.rename(columns = {'runtime' : 'max_runtime'},inplace=True)
# We can find the median movie runtime across different years
imdb_df.groupby('release_year')['runtime'].median().reset_index()
# Group the data frame by year and item and extract a number of stats from each group
imdb_df.groupby(['release_year', 'original_language'],as_index=False).agg({'runtime':"mean",'revenue': "sum",'imdb_id':"count"}).sort_values(by='release_year').rename(columns ={'runtime':'avg_runtime','revenue':'total_revenue','imdb_id':'num_movies'})
```

### Activity 4
 Your friend is a film enthusiast and has decided to explore movies in languages other than English. He wants recommendations on which movies to watch next.

Using the dataset provided, your task is to find out the language with the highest mean popularity rating and then list the top 5 movies in that language.

Can you help your fellow film enthusiast friend by completing this task and providing them with recommendations for non-English movies?

Note: idxmax(): This function is then applied to the resulting Series to get the index (in this case, the language) where the maximum mean popularity occurs.

```python
highest_mean_language = imdb_df.groupby('original_language')['popularity'].mean().idxmax()

# Filter movies in the language with the highest mean popularity rating
top_language_movies = imdb_df[imdb_df['original_language'] == highest_mean_language]
print(top_language_movies)
# Sort the top language movies by popularity and get the top 5
top_5_movies = top_language_movies.sort_values(by='popularity', ascending=False).head(5)

# Print the language with the highest mean popularity rating
print(f"Language with the highest mean popularity rating: {highest_mean_language}")

# Print the top 5 movies in that language
print("\nTop 5 Movies:")
print(top_5_movies[['title', 'popularity']])
```
