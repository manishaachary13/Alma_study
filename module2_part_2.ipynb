````python
# Here's an example program for each of the functions with explanations:

#1. `dropna()`

import pandas as pd

# Create a DataFrame with missing values
data = {'Name': ['Alice', 'Bob', 'Charlie', None],
        'Age': [25, 30, None, 22],
        'Salary': [50000, 60000, 75000, None]}
df = pd.DataFrame(data)

# Drop rows with missing values
df.dropna(inplace=True)

print("DataFrame after dropping rows with missing values:")
print(df)

# Explanation: This program creates a DataFrame with missing values and uses the `dropna()` function to remove rows containing
# any missing values.
````

````python
#2. `fillna()`

import pandas as pd

# Create a DataFrame with missing values
data = {'Name': ['Alice', 'Bob', 'Charlie', None],
        'Age': [25, 30, None, 22],
        'Salary': [50000, 60000, 75000, None]}
df = pd.DataFrame(data)

# Fill missing values in the 'Age' column with the mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

print("DataFrame after filling missing values:")
print(df)
print(df.dropna())
# Explanation: This program fills missing values in the 'Age' column with the mean value of that column using the `fillna()` function.
````

````python
#3. `replace()`
import pandas as pd
# Create a DataFrame with a column containing specific values
data = {'Category': ['A', 'B', 'A', 'C', 'B']}
df = pd.DataFrame(data)

# Replace 'A' with 'New' in the 'Category' column
df['Category'].replace('A', 'New', inplace=True)
# replacing more than one value
replace_dict={'A':'AAA','B':'BBB','C':'CCC'}
df['Category'].replace(replace_dict, inplace=True)

print("DataFrame after replacing values:")
print(df)

# Explanation: This program replaces occurrences of 'A' with 'New' in the 'Category' column using the `replace()` function.
````

````python
#4. `loc()`

import pandas as pd

# Create a DataFrame with labeled rows and columns
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'Salary': [50000, 60000, 75000]}
df = pd.DataFrame(data, index=['A', 'B', 'C'])
print(df)
# Select the row with index 'B' and the 'Age' and 'Salary' columns

selected_dat = df.loc['B', ['Age', 'Salary']]
selected_data = df.loc[['B','C'], ['Age', 'Salary']]

print("Selected data using loc:")
print(selected_dat)
print(selected_data)
# Explanation: This program uses the `loc()` function to select specific rows and columns from a DataFrame based on labels.
````

````python
### 5. `iloc()`
import pandas as pd
# Create a DataFrame with labeled rows and columns
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'Salary': [50000, 60000, 75000]}
df = pd.DataFrame(data, index=['A', 'B', 'C'])
# Select the first row and the first two columns using integer positions
selected_data = df.iloc[0, :2]
print("Selected data using iloc:")
print(selected_data)

# Explanation: This program uses the `iloc()` function to select specific rows and columns from a DataFrame based on integer positions.
````

````python
# 6. `merge()`

import pandas as pd
# Create two DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Salary': [50000, 60000, 70000]})
# Merge DataFrames based on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print("Merged DataFrame:")
print(merged_df)
# Explanation: This program uses the `merge()` function to combine two DataFrames based on a common column ('ID' in this case).
````

````python
# defines a function called revised_profit that takes two arguments,
# revenue and budget, and returns a new profit value if the revenue is greater than 0, or NaN if the revenue is less than or equal to 0
def revised_profit(revenue,budget):
    if revenue > 0:
        new_profit = revenue - budget
    else:
        new_profit = np.nan
    return new_profit

imdb_df['new_profit'] = imdb_df.apply(lambda x: revised_profit(x['revenue'],x['budget']),axis=1)
imdb_df['new_profit'].head()

#
````

````python
# 4.Use Pandas to analyze the dataset and provide a report summarizing your findings.
# Be sure to include any data cleaning or transformation steps you took to arrive at your conclusions.

# import numpy, pandas and drive
import numpy as np
import pandas as pd
from google.colab import drive
# mounting the drive into colab
drive.mount('/content/drive')
# set file path
path = '/content/drive/MyDrive/Colab Notebooks/movies.csv'
# read file and store in a dataframe
movies = pd.read_csv(path)
movies
# 1.What is the average box office revenue per year for the movies in the dataset?
average_revenue_year=movies.groupby('Release Year')['Box Office'].mean()
# 2.Which genres are the most popular and how many movies have been released in each genre?
movies_per_genre=movies['Genre'].value_counts()
# 3.Who are the top 5 directors in terms of box office revenue?
directors_revenue=movies.groupby('Director')['Box Office'].sum()
sorted_directors=directors_revenue.sort_values(ascending=False)
print("Average revenue per year:",average_revenue_year)
print("Movies in each geres:",movies_per_genre)
print("Top 5 directors based on their box office collection:\n",sorted_directors.head())
````
