```python
#Imports packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from matplotlib.ticker import MaxNLocator

df_layoffs=pd.read_csv("industry.csv")
plt.figure(figsize= (8, 6))
industry_val = df_layoffs.groupby('company')['total_laid_off'].sum().sort_values(ascending = False).head(10)
industry_val.plot(label="", kind='bar')
plt.show()
```

```python
# Group the data by industry and sum the total laid off employees
industry_val = df_layoffs.groupby('industry')['total_laid_off'].sum().sort_values(ascending=False).head()

# create the pie chart and display the labels and values inside the pie
plt.figure(figsize=(8, 6))
plt.pie(industry_val, labels=industry_val.index, autopct='%1.1f%%')
plt.title('Laid Off Employees by Industry')
plt.show()
```

```python
# convert date column to datetime object
df_layoffs['date'] = pd.to_datetime(df_layoffs['date'])

# select data for one-year duration starting from January 1st, 2022
start_date = pd.Timestamp('2022-01-01')
end_date = start_date + pd.DateOffset(years=1)
df_one_year = df_layoffs.loc[(df_layoffs['date'] >= start_date) & (df_layoffs['date'] < end_date)]

# plot the selected data
df_date = df_one_year.groupby('date')['total_laid_off'].sum()
plt.figure(figsize=(10, 4))
plt.plot(df_date.index, df_date.values)
plt.xlabel('Date')
plt.ylabel('Total Laid Off')
plt.title('Laid Off Trend for 2022')
plt.xticks(rotation=45)
# set the format of the x-axis labels to show Month-Year
date_fmt = mdates.DateFormatter('%b-%Y')
plt.gca().xaxis.set_major_formatter(date_fmt)

# Use MaxNLocator to reduce the number of xticks
locator = MaxNLocator(nbins=10)
plt.gca().xaxis.set_major_locator(locator)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MaxNLocator

# convert date column to datetime object
df_layoffs['date'] = pd.to_datetime(df_layoffs['date'])

# filter data to only include 2020 and 2022
df_filtered = df_layoffs[(df_layoffs['date'].dt.year == 2020) | (df_layoffs['date'].dt.year == 2022)]

# group data by year and calculate total layoffs
df_filtered['year'] = df_filtered['date'].dt.year
df_yearly = df_filtered.groupby(['year', 'date'])['total_laid_off'].sum().reset_index()

# create subplots and plot the data for each year in separate charts
fig, axs = plt.subplots(ncols=2, figsize=(14, 8))
for i, year in enumerate(df_yearly['year'].unique()):
    df_year = df_yearly.loc[df_yearly['year'] == year]
    axs[i].plot(df_year['date'], df_year['total_laid_off'])
    axs[i].set_xlabel('Date')
    axs[i].set_ylabel('Total Laid Off')
    axs[i].set_title(f'Laid Off Trend for {year}')
    axs[i].xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))
    axs[i].tick_params(axis='x', rotation=45)
    locator = MaxNLocator(nbins=10)
    axs[i].xaxis.set_major_locator(locator)

# set y-axis limit to 0-14000 for each subplot
for ax in axs:
    ax.set_ylim([0, 14000])

plt.show()
```

```python
automobile=pd.read_csv("automobile.csv")
# find out the missing price rows
automobile.loc[automobile['price']=='?']
```

```python
# replace ? with nan
automobile.replace('?',np.nan,inplace=True)
# drop missing value data
automobile.dropna(subset=['price'],inplace=True)
```

```python
# replace missing values in normalized-losses column with mean value
automobile['normalized-losses'].fillna(automobile['normalized-losses'].astype(float).mean(),inplace=True)
automobile.head(10)
automobile.dtypes
```

```python
# Convert data types from object to float
automobile[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']] = automobile[['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']].astype(float)
automobile.dtypes
```

```python
duplicated_data=automobile[automobile.duplicated()]
duplicated_data
automobile.drop_duplicates(inplace=True)
automobile
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a box plot of the price feature
sns.boxplot(x="price", data=automobile)

# Display the plot
plt.show()
```

```python
# Dealing with outliers
automobile= automobile[(automobile['price'] >= automobile['price'].quantile(0.05)) &
        (automobile['price'] <= automobile['price'].quantile(0.95))]
automobile
```

```python
sns.histplot(automobile['price'], kde=True, color='black', bins=20)
plt.title("Histogram fore price of automobiles with kde True")
plt.show()
automobile.columns
```

```python
# Scatterplot of horsepower vs. price colored by fuel-type:

sns.scatterplot(x='horsepower', y='price', hue='engine-location', data=automobile)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(automobile.corr(),annot=True)
```

```python
import seaborn as sns
fig, ax = plt.subplots()
sns.barplot(x='horsepower', y='price', data=automobile)
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar Plot')
plt.show()
```

```python
import seaborn as sns
custom_palette = sns.color_palette(['#ff7f0e', '#1f77b4', '#2ca02c'])
fig, ax = plt.subplots()
sns.barplot(x='horsepower', y='price', data=automobile, palette=custom_palette)
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Bar Plot')
plt.show()
```
