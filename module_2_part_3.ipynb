````python
import re

def split_into_sentences(text):
    # Define the regular expression pattern for sentence splitting
    sentence_pattern = re.compile(r'(?<=[.!?])\s+')

    # Use re.split() to split the text into sentences
    sentences = sentence_pattern.split(text)

    return sentences

# Take input from the user
user_input = input("Enter a string of text: ")

# Split the input into sentences
result_sentences = split_into_sentences(user_input)

# Print out the list of sentences
print("List of Sentences:")
for i, sentence in enumerate(result_sentences, start=1):
    print(f"{i}. {sentence}")
````

````python
'''Age Calculator'''
from datetime import datetime

def calculate_age(birthdate):
    birthdate = datetime.strptime(birthdate, "%Y-%m-%d")
    today = datetime.now()
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age

# Example usage:
birthdate_input = input("Enter your birthdate (YYYY-MM-DD): ")
age = calculate_age(birthdate_input)
print(f"You are {age} years old.")
````

````python
'''Days until next Birthday'''
from datetime import datetime, timedelta

def days_until_next_birthday(birthdate):
    birthdate = datetime.strptime(birthdate, "%Y-%m-%d")
    today = datetime.now()
    next_birthday = datetime(today.year, birthdate.month, birthdate.day)

    if today > next_birthday:
        next_birthday = datetime(today.year + 1, birthdate.month, birthdate.day)

    days_until_birthday = (next_birthday - today).days
    return days_until_birthday

# Example usage:
birthdate_input = input("Enter your birthdate (YYYY-MM-DD): ")
days_until_bday = days_until_next_birthday(birthdate_input)
print(f"Days until your next birthday: {days_until_bday}")
````

````python
'''Given a text dataset, extract all email addresses using regular expressions.'''
import re

def extract_emails(text):
    email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    emails = re.findall(email_pattern, text)
    return emails

# Example Usage:
text_data = "Contact us at john.doe@example.com or info@company.com for more information."
result_emails = extract_emails(text_data)
print("Extracted Emails:", result_emails)
````

````python
'''Given a text dataset, identify and extract all occurrences of dates using regular expressions.'''
import re

def extract_dates(text):
    date_pattern = r'\b\d{4}-\d{2}-\d{2}\b'
    dates = re.findall(date_pattern, text)
    return dates

# Example Usage:
text_data = "The meeting is scheduled for 2022-01-15 and the report is due on 2022-02-28."
result_dates = extract_dates(text_data)
print("Extracted Dates:", result_dates)
````

````python
'''Given a dataset containing textual information, extract all phone numbers using regular expressions.'''
import re

def extract_phone_numbers(text):
    phone_pattern = r'\b\d{3}[-.\s]?\d{3}[-.\s]?\d{4}\b'
    phone_numbers = re.findall(phone_pattern, text)
    return phone_numbers

# Example Usage:
text_data = "Contact us at 123-456-7890 or 987.654.3210 for assistance."
result_phone_numbers = extract_phone_numbers(text_data)
print("Extracted Phone Numbers:", result_phone_numbers)
````

````python
'''Create a program to tokenize a given text into individual words using regular expressions.'''
import re

def tokenize_text(text):
    word_pattern = r'\b\w+\b'
    words = re.findall(word_pattern, text)
    return words

# Example Usage:
text_data = "This is a sample sentence for tokenization."
result_words = tokenize_text(text_data)
print("Tokenized Words:", result_words)
````

````python
'''Create a program to extract keywords associated with positive or negative sentiment from a given text dataset.'''
import re

def extract_sentiment_keywords(text, sentiment='positive'):
    if sentiment == 'positive':
        pattern = r'\b(?:good|excellent|positive|awesome)\b'
    elif sentiment == 'negative':
        pattern = r'\b(?:bad|poor|negative|awful)\b'
    else:
        return []

    keywords = re.findall(pattern, text, flags=re.IGNORECASE)
    return keywords

# Example Usage:
text_data = "The movie was excellent! I had a bad experience with the service though."
result_positive_keywords = extract_sentiment_keywords(text_data, sentiment='positive')
result_negative_keywords = extract_sentiment_keywords(text_data, sentiment='negative')

print("Positive Keywords:", result_positive_keywords)
print("Negative Keywords:", result_negative_keywords)
````

````python
import requests
from bs4 import BeautifulSoup
import json
import csv

# Your HTML content
url = "https://www.imdb.com/chart/top/"

# Make a request to the webpage and get the response
response = requests.get(url, headers={'User-Agent': "Mozilla/5.0"})

# Create a BeautifulSoup object to parse the HTML content of the webpage
soup = BeautifulSoup(response.text, "html.parser")


# Find the script tag with the __NEXT_DATA__ id
script_tag = soup.find('script', id='__NEXT_DATA__')

# Extract the JSON data from the script tag
json_data = json.loads(script_tag.string)

# Navigate to the relevant data
chart_titles = json_data['props']['pageProps']['pageData']['chartTitles']['edges']

# Initialize lists to store movie data
titles = []
release_years = []
ratings = []

# Iterate through the movie data and extract the relevant information
for chart_title in chart_titles:
    movie = chart_title['node']
    title_chinese = movie['titleText']['text']
    release_year = movie['releaseYear']['year']
    rating = movie['ratingsSummary']['aggregateRating']

    titles.append(title_chinese)
    release_years.append(release_year)
    ratings.append(rating)

# Export the data to a CSV file
with open('top_movies.csv', 'w', newline='') as csvfile:
    fieldnames = ['Title (Chinese)', 'Release Year', 'Rating']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for i in range(len(titles)):
        writer.writerow({'Title (Chinese)': titles[i], 'Release Year': release_years[i], 'Rating': ratings[i]})
````

You are tasked with retrieving the list of courses offered by a university from their website. You are given the following HTML code for the university's courses page:html_doc = """



Using Beautiful Soup, write a Python program that extracts the names of all the courses offered by the university and prints them to the console.

Instructions:

1. Create a new Python file and import the necessary Beautiful Soup library.

1. Open the HTML file and read its contents.

1. Parse the HTML using Beautiful Soup.

1. Find all the ul tags in the HTML.

1. Loop through each ul tag and find all the li tags within it.

1. Loop through each li tag and extract its text.

1. Print the text of each li tag to the console.

````python
from bs4 import BeautifulSoup

html_doc = """
<html>
<head>
	<title>University Courses</title>
</head>
<body>
	<h1>University Courses</h1>
	<h2>Computer Science</h2>
	<ul>
		<li>Introduction to Programming</li>
		<li>Data Structures and Algorithms</li>
		<li>Operating Systems</li>
	</ul>
	<h2>Mathematics</h2>
	<ul>
		<li>Calculus I</li>
		<li>Linear Algebra</li>
		<li>Differential Equations</li>
	</ul>
	<h2>History</h2>
	<ul>
		<li>World History I</li>
		<li>World History II</li>
		<li>US History</li>
	</ul>
</body>
</html>
"""

# Parse the HTML using Beautiful Soup
soup = BeautifulSoup(html_doc, 'html.parser')

# Find all the h2 tags (subject titles) in the HTML
subject_titles = soup.find_all('h2')

# Loop through each subject title
for title in subject_titles:
    # Print the subject title
    print(f'{title.text}:', end=' ')

    # Find the ul tag following the subject title
    ul_tag = title.find_next('ul')

    # Find all the li tags within the ul tag
    li_tags = ul_tag.find_all('li')

    # Extract and print the text of each li tag
    courses = [li.text for li in li_tags]
    print(', '.join(courses))
````

````python
import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page for Python programming language
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content using Beautiful Soup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the main content of the page (usually within the <p> tags)
    paragraphs = soup.find_all('p')

    # Print the text of each paragraph to the console
    for paragraph in paragraphs:
        print(paragraph.text)

else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
````

````python
import requests
from bs4 import BeautifulSoup
import csv
from prettytable import PrettyTable

def scrape_github_trending_repos(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        repo_list = soup.select('article.Box-row')

        repo_data = []

        for repo in repo_list[:10]:  # Considering the top 10 repositories
            repo_name = repo.select_one('h1.h3 a').text.strip()
            repo_owner = repo.select_one('span.f4.text-normal').text.strip()
            repo_url = 'https://github.com' + repo.select_one('h1.h3 a')['href']

            repo_data.append({
                'Repository Name': repo_name,
                'Owner': repo_owner,
                'Repository URL': repo_url
            })

        return repo_data
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

def write_to_csv(repo_data):
    with open('python_trending_repos.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['Repository Name', 'Owner', 'Repository URL'])
        writer.writeheader()

        for repo in repo_data:
            writer.writerow(repo)

def display_table(repo_data):
    table = PrettyTable()
    table.field_names = ['Repository Name', 'Owner', 'Repository URL']

    for repo in repo_data:
        table.add_row([repo['Repository Name'], repo['Owner'], repo['Repository URL']])

    print(table)

def main():
    github_url = "https://github.com/trending/python?since=daily"
    repo_data = scrape_github_trending_repos(github_url)

    if repo_data:
        write_to_csv(repo_data)
        display_table(repo_data)

if __name__ == "__main__":
    main()
````

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Function to scrape COVID-19 data
def scrape_covid_data(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extracting data for each country
        country_data = []
        table = soup.find('table', {'id': 'main_table_countries_today'})
        rows = table.find_all('tr')[8:]  # Skip header rows
        for row in rows:
            columns = row.find_all(['th', 'td'])
            country_name = columns[1].text.strip()
            total_cases = columns[2].text.strip().replace(',', '')
            total_deaths = columns[4].text.strip().replace(',', '')
            total_recovered = columns[6].text.strip().replace(',', '')
            active_cases = columns[8].text.strip().replace(',', '')

            country_data.append({
                'Country': country_name,
                'Total Cases': total_cases,
                'Total Deaths': total_deaths,
                'Total Recovered': total_recovered,
                'Active Cases': active_cases
            })

        return country_data
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None

# Function to clean and process data
def clean_and_process_data(data):
    df = pd.DataFrame(data)
    df = df.apply(pd.to_numeric, errors='coerce')  # Convert columns to numeric, handle non-numeric values as NaN
    df = df.dropna()  # Remove rows with missing values
    df = df.astype({'Total Cases': 'int', 'Total Deaths': 'int', 'Total Recovered': 'int', 'Active Cases': 'int'})

    return df

# Function to visualize the data
def visualize_data(df):
    # Example: Plotting top 10 countries by Total Cases
    top_10_countries = df.nlargest(10, 'Total Cases')
    plt.figure(figsize=(10, 6))
    plt.bar(top_10_countries['Country'], top_10_countries['Total Cases'])
    plt.xlabel('Country')
    plt.ylabel('Total Cases')
    plt.title('Top 10 Countries by Total Cases')
    plt.xticks(rotation=45)
    plt.show()

# Main function
def main():
    covid_url = "https://www.worldometers.info/coronavirus/"
    covid_data = scrape_covid_data(covid_url)

    if covid_data:
        df = clean_and_process_data(covid_data)
        visualize_data(df)
        df.to_csv('covid19_data.csv', index=False)

if __name__ == "__main__":
    main()
````
